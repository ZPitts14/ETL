```python
# Dependencies
import pandas as pd
```

```python
# Store filepath in a variable
nba_data = "Resources/nba_season_data_07_17.csv"
win_shares = "Resources/win_shares.csv"
```

```python
# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
nba_data_df = pd.read_csv(nba_data, encoding="ISO-8859-1")
win_shares_df = pd.read_csv(win_shares, encoding="ISO-8859-1")
```

```python
# Show just the header
nba_data_df.head()
```

```python
nba_data_df['player_height'] = nba_data_df['player_height'].astype(float)
```

```python
nba_data_df['player_height'] = nba_data_df['player_height'].apply(lambda x : x*0.393071)
```

```python
nba_data_df = nba_data_df.round(2)
```

```python
mia_query_df = nba_data_df.query('team_abbreviation == "MIA" & season == "2012-13"')
mia_query_df.head()
```

```python
sas_query_df = nba_data_df.query('team_abbreviation == "SAS" & season == "2013-14"')
sas_query_df.head()
```

```python
gsw_query_df = nba_data_df.query('team_abbreviation == "GSW" & season == "2014-15"')
gsw_query_df.head()
```

```python
win_shares_df.head()
```

```python
finals_df = win_shares_df.query('Year == "2013" | Year == "2014" | Year == "2015"')
finals_df
```

```python
# Export file as a CSV, without the Pandas index, but with the header
finals_df.to_csv("finals.csv", index=False, header=True)
mia_query_df.to_csv("mia.csv", index=False, header=True)
sas_query_df.to_csv("sas.csv", index=False, header=True)
gsw_query_df.to_csv("gsw.csv", index=False, header=True)
```
